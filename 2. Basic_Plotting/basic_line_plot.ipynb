# Basic Line Plotting with Matplotlib #

### 1. Importing libraries ###

```python
from matplotlib import pyplot as plt
import pandas as pd
```

### 2. Loading Dataset ###

```python
dataURL = "https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv"
iris = pd.read_csv(dataURL)
iris.head()
```

### 3. Create a Basic Line Plot ###

![Iris Flower](iris.png)

```python
plt.plot(iris['sepal_length'])
plt.title("Sepal Length of Iris")
plt.xlabel('Sample Index')
plt.ylabel('Sepal length (cm)')
plt.show()
```

```python
plt.plot(iris['sepal_length'], linestyle='-', color='orange', marker='.')
plt.title('Customized Sepal Length Plot')
plt.xlabel('Sample Index')
plt.ylabel('Sepal Length (cm)')
plt.savefig("Iris Sepal Length Plot")
plt.show()
```
