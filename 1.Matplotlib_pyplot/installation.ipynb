### 2. Import `pyplot` from `matplotlib` ###

```python
from matplotlib import pyplot as plt 
```

### 3. Generate a simple plot using `matplotlib` ###

```python
# Sample Data
x = [1,2,3,4,5]
y = [2,4,6,7,11]

# Create a plot
plt.plot(x,y)

# Display the plot
plt.show()
```

### 4. Add a title and axis labels to the plot ###

```python
# Sample Data
x = [1,2,3,4,5]
y = [2,4,6,7,11]

# Title and labels
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.title("Sample Data Plot")

# Create a plot
plt.plot(x,y)

# Display the plot
plt.show()
```

### 5. Customize the graph ###

```python
# Sample Data
x = [1,2,3,4,5]
y = [2,4,6,7,11]

# Title and labels
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.title("Sample Data Plot")

# Create a plot and add marker for points and change line style
plt.plot(x,y,marker='o',linestyle='--',color='green')

# Create a legend
plt.legend(["Data Point"])

# Display the plot
plt.show()
```

### 6. Adding two graphs ###

```python
# Sample Data
x = [1,2,3,4,5]
y = [2,4,6,7,11]

# Title and labels
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.title("Sample Data Plot")

# Create a plot and add marker for points and change line style
plt.plot(x,y,marker='o',linestyle='--',color='green')

# Create a legend
plt.legend(["Data Point"])

# Display the plot
# plt.show()

# Save as image (don't use both show and save together)
plt.savefig("plot1.png")
```

```python
# Sample Data
x1 = [1,2,3,4,5]
y1 = list(map(lambda x : x**2 + 3*x + 8, x1))

x2 = [1,2,3,4,5]
y2 = list(map(lambda x : x**3 +  7, x2))

# Title and labels
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.title("Sample Data Plot")

# Create a plot and add marker for points and change line style
plt.plot(x1,y1,marker='o',linestyle='--',color='green')
plt.plot(x2,y2,marker='x',color='red')

# Create a legend
plt.legend(["Data Point 1","Data Point 2"])

# Display the plot
plt.show()
```
