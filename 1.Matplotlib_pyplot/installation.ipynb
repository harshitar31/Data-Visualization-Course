### 2. Import `pyplot` from `matplotlib` ###
Once Matplotlib is installed, you can import its `pyplot` module — commonly used for plotting — using the following command:

```python
from matplotlib import pyplot as plt 
```

### 3. Generate a simple plot using `matplotlib` ###
Once Matplotlib is installed, you can start creating simple plots. Below is a minimal example of how to plot a line chart using two lists of values:

- `x`: the horizontal axis values
- `y`: the vertical axis values

The `plt.plot()` function creates the line plot, and `plt.show()` displays it in the notebook.

```python
# Sample Data
x = [1,2,3,4,5]
y = [2,4,6,7,11]

# Create a plot
plt.plot(x,y)

# Display the plot
plt.show()
```

### 4. Add a Title and Axis Labels to the Plot ###

To make your plots more informative, it's a good practice to include a title and axis labels. This helps others (and future you!) quickly understand what the chart represents.

- `plt.title()` adds a title to the plot.
- `plt.xlabel()` sets the label for the x-axis.
- `plt.ylabel()` sets the label for the y-axis.

Below is an example where we use these functions to enhance the readability of a simple line chart.


```python
# Sample Data
x = [1,2,3,4,5]
y = [2,4,6,7,11]

# Title and labels
plt.xlabel("X-Label")
plt.ylabel("Y-Label")
plt.title("Sample Data Plot")

# Create a plot
plt.plot(x,y)

# Display the plot
plt.show()
```
