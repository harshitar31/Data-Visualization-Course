### Multi Plots


In this notebook, we explore how to create multiple plots within a single figure using Matplotlib.

This is useful when comparing data side by side or presenting different views of the same dataset.

We begin by importing the required libraries:

```python
from matplotlib import pyplot as plt
import numpy as np
```

### Creating Multiple Subplots with `plt.subplot()`

This example demonstrates how to plot multiple graphs in the same figure using `plt.subplot()` â€” a convenient way to compare different data visualizations side by side or top to bottom.


**Figure & Subplot Layout**
- `plt.figure(figsize=(width, height))`: Sets the overall figure size in inches.
- `plt.subplot(nrows, ncols, index)`: Creates a grid of subplots and selects the subplot at the given index (starts from 1).
  - Example: `plt.subplot(2, 1, 1)` means 2 rows, 1 column, selecting the 1st plot.
- Each subplot can have its own:
  - `plt.plot()`
  - `plt.title()`
  - `plt.xlabel()` and `plt.ylabel()`
  - `plt.legend()`


**Styling**
- `plt.style.use('style-name')`: Applies a consistent theme across all plots.
  - Example: `'seaborn-v0_8-darkgrid'` adds a dark grid background.
  - Use `plt.style.available` to list all available styles.
- Style affects grid, background, font, and color palette.


**Labels & Legends**
- `plt.title()`, `plt.xlabel()`, `plt.ylabel()` can be applied to each subplot individually.
- `plt.legend()` helps distinguish lines on each plot. Use `label=` inside `plot()` and call `legend()` to display it.
- You can place the legend with `loc='upper right'`, `'lower left'`, etc.


**Tip: Managing Overlap**
- Use `plt.tight_layout()` after all subplots to auto-adjust spacing and avoid overlapping text (especially for titles and axis labels).


```python
plt.style.use('seaborn-v0_8-darkgrid')
plt.figure(figsize=(12,5))
x = np.linspace(0,10,100)
y1 = np.sin(x)
y2 = np.tan(x)
plt.subplot(2,1,1)
plt.plot(x,y1,'rs',label='$y1=sin(x)$')
plt.xlabel('x')
plt.ylabel('$sin(x)$')
plt.title('$sin(x)$')
plt.legend(loc='upper right')
plt.subplot(2,1,2)
plt.plot(x,y2, 'b-', label='$y2=tan(x)$')
plt.xlabel('x')
plt.ylabel('$tan(x)$')
plt.title('$tan(x)$')
plt.legend(loc='upper right')
```

In this example, we use `plt.subplots()` to create a grid of subplots in a single figure. This approach is more flexible and scalable than using `plt.subplot()`.

- `plt.subplots(nrows, ncols, figsize=(width, height))`: Creates a grid of subplots and returns a figure and an array of Axes objects.
  - Here, a 2Ã—3 grid of subplots is created with a custom figure size of 20Ã—10 inches.
- `ax[row, col].plot(...)`: Plots data on a specific subplot using the returned Axes array.
- `ax[row, col].set_title()` is used to assign a title to each subplot.
- `plt.tight_layout()` automatically adjusts spacing between plots to avoid label or title overlap.

Each subplot in this figure shows one of the following mathematical functions:
- `sin(x)`
- `cos(x)`
- `tan(x)`
- `sinh(x)`
- `cosh(x)`
- `tanh(x)`

This layout is particularly useful when visualizing multiple functions or comparisons side by side in a clean and organized grid.


```python
x = np.arange(-6,6,0.05)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)
y4 = np.sinh(x)
y5 = np.cosh(x)
y6 = np.tanh(x)
fig, ax1 = plt.subplots(nrows=2, ncols=3, figsize=(20,10))
ax1[0,0].plot(x, y1, 'tab:orange')
ax1[0,0].set_title('$sin(x)$')
ax1[0,1].plot(x, y2, 'tab:green')
ax1[0,1].set_title('$cos(x)$')
ax1[0,2].plot(x, y3, 'b-')
ax1[0,2].set_title('$tan(x)$')
ax1[1,0].plot(x, y4, 'r-')
ax1[1,0].set_title('$sinh(x)$')
ax1[1,1].plot(x, y5, 'g-')
ax1[1,1].set_title('$cosh(x)$')
ax1[1,2].plot(x, y6, 'm-')
ax1[1,2].set_title('$tanh(x)$')
plt.tight_layout()
```

### Plotting Multiple Functions with a Loop

Instead of writing multiple `plt.plot()` calls manually, we can use a loop to plot several functions on the same axes. This approach is efficient when dealing with similar computations or repetitive plotting.

#### Features Used:
- `np.power(x, n)` computes element-wise exponentiation.
- A list `y` contains multiple power functions: $x^2$, $x^3$, and $x^4$.
- The `for` loop iterates through each function and plots it on the same figure.
- `plt.figure(figsize=(15, 10))` increases the size of the plot for better visibility.

This is useful when visualizing a family of related functions or transformations.

```python
x=np.arange(-10,10,0.05)
y=[np.power(x,2),np.power(x,3),np.power(x,4)]
plt.figure(figsize=(15,10))
for i in range(len(y)):
      plt.plot(x,y[i],label=f"$x^{i+2}$")
plt.legend(loc='upper center',fontsize='30')
plt.savefig("xsq,xcub,xpower4")
```
