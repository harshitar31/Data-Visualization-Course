## Line Styles, Axes, Labels ##

In this section, we explore how to customize line plots to improve readability and aesthetics.

We’ll cover:
- Different line styles and how to apply them
- Adding and customizing axis labels and chart titles
- Enhancing plots with markers and colors

We begin by importing the necessary libraries:

```python
from matplotlib import pyplot as plt 
import pandas as pd 
import numpy as np
```

### Loading the Iris Dataset

We’ll use the well-known **Iris dataset** to demonstrate line styles and labeling techniques.

- `pd.read_csv()` loads the dataset directly from a URL.
- `iris.head()` displays the first few rows to understand the data structure.

```python
url = 'https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv'
iris = pd.read_csv(url)
iris.head()
```

###  Basic Line Plot with Custom Styling and Labels

Let’s create a line plot of the **sepal length** feature from the Iris dataset with added styling:

- `linestyle='-'` draws a solid line.
- `color='green'` sets the line color.
- `marker='.'` places a dot at each data point.
- `plt.title()`, `plt.xlabel()`, and `plt.ylabel()` add descriptive labels to the plot.

```python
plt.plot(iris['sepal_length'],linestyle='-',color='green',marker='.')
plt.title('Sepal Length of Iris Samples')
plt.xlabel('Index')
plt.ylabel('Sepal Length $(cm)$')
plt.show()
```

### Matplotlib Styling Options 

Customize your plots using these powerful options:

**Colors**
- `'b'`, `'g'`, `'r'`, `'c'`, `'m'`, `'y'`, `'k'`, `'w'` = basic colors
- Full color names like `'blue'`, `'green'`, `'orange'`, `'purple'`, etc.
- Hex codes (e.g., `'#FF5733'`) and RGB tuples like `(0.1, 0.2, 0.5)` are supported.


**Line Styles**
- `'-'` = solid (default)  
- `'--'` = dashed  
- `'-.'` = dash-dot  
- `':'` = dotted  
- `'None'` or `''` = no line (useful with markers only)


**Markers**
- `'.'` = point  
- `'o'` = circle  
- `'s'` = square  
- `'^'` = triangle up  
- `'v'` = triangle down  
- `'*'` = star  
- `'+'` = plus  
- `'x'` = x  
- `'D'` = diamond  
- `'h'` = hexagon  
- `'None'` or `''` = no marker

**Marker Customization**
- `markersize` = size of the marker (e.g., `markersize=10`)  
- `markerfacecolor` = fill color inside the marker  
- `markeredgecolor` = outline color of the marker  


**Fonts and Labels**
- `fontsize` = control text size (e.g., `fontsize=12`)  
- `fontweight='bold'` = makes title or label text bold  
- `fontfamily='serif'` or `'sans-serif'` = changes font style  


**Axis Limits & Ticks**
- `plt.xlim(min, max)` / `plt.ylim(min, max)` = set visible axis range  
- `plt.xticks(np.arange(start, stop, step))` = customize x-axis ticks  
- `plt.yticks(np.arange(start, stop, step))` = customize y-axis ticks  


**Grid**
- `plt.grid(True)` = show grid  
- `plt.grid(axis='x')` or `'y'` = show gridlines only on one axis  


**Saving Plots**
- `plt.savefig("filename.png")` = save the plot as an image  
- Save **before** `plt.show()` for best results  
- File types: `.png`, `.jpg`, `.pdf`, `.svg`, etc.

These options allow you to create clean, professional-quality plots for analysis and presentation.


### Advanced Line Plot Customization

This example demonstrates how to take full control of a Matplotlib plot’s appearance:

- `markersize` and `markerfacecolor` fine-tune the marker look.
- `fontsize` and `fontweight` improve label readability.
- `plt.xlim()` and `plt.ylim()` manually set axis ranges.
- `plt.xticks()` and `plt.yticks()` control tick spacing and appearance.
- `plt.grid(True)` enables gridlines for better visual alignment.
- `plt.savefig()` saves the plot to an image file — make sure to call this **before** `plt.show()`.

```python
plt.plot(iris['sepal_width'],linestyle='-',color='green',marker='.',markersize=10,markerfacecolor='darkgreen')
plt.title('Sepal Width of Iris Samples',fontsize=16, fontweight='bold')
plt.xlabel('Index', fontsize=12)
plt.ylabel('Sepal Width $(cm)$',fontsize=12)
plt.xlim(0,150)
plt.ylim(0,5)
plt.xticks(np.arange(0,150,20))
plt.yticks(np.arange(0,5,0.5))
plt.grid(True)
# plt.show()
plt.savefig("sepal width of iris.png")
```
