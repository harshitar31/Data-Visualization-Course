## Line Styles, Axes, Labels ##

```python
from matplotlib import pyplot as plt 
import pandas as pd 
import numpy as np
```

```python
url = 'https://raw.githubusercontent.com/uiuc-cse/data-fa14/gh-pages/data/iris.csv'
iris = pd.read_csv(url)
iris.head()
```

```python
plt.plot(iris['sepal_length'],linestyle='-',color='green',marker='.')
plt.title('Sepal Length of Iris Samples')
plt.xlabel('Index')
plt.ylabel('Sepal Length $(cm)$')
plt.show()
```

```python
plt.plot(iris['sepal_width'],linestyle='-',color='green',marker='.',markersize=10,markerfacecolor='black')
plt.title('Sepal Width of Iris Samples',fontsize=16, fontweight='bold')
plt.xlabel('Index', fontsize=12)
plt.ylabel('Sepal Width $(cm)$',fontsize=12)
plt.xlim(0,150)
plt.ylim(0,5)
plt.xticks(np.arange(0,150,20))
plt.yticks(np.arange(0,5,0.5))
plt.grid(True)
plt.show()
```
